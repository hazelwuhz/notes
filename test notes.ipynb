##### Lecture 0: set theory

Union: $A\cup B$	Intersection: $A\cap B$

Difference: $A-B$, $A\backslash B$

Symmetric Difference: $A\triangle B$

$\wp(S)=\{T|T\subseteq S\}$ is the power set of $S$, the set of all subsets of $S$.

The cardinality $|S|$ of a set $S$ is the number of elements it contains.

​	$|\mathbb{N}|=\mathbb{Z}|=\aleph_0$

Cantor’s Theorem: If $S$ is a set, then $|S| < |\wp(S)|$.

##### Lecture 2: indirect proofs

A proposition is a statement that is either true or false.

An implication is a statement of the form “If $P$ is true, then $Q$ is true.”
$\text{antecedent}\to\text{consequent}$

proof by contradiction:

> Assume for the sake of contradiction that …
>
> We have reached a contradiction, so our assumption must have been wrong. Therefore, …

Proof by contrapositive:

> We will prove the contrapositive of this statement, that if …, then …

##### Lecture 3: propositional logic

Propositional connectives link propositions into larger propositions.

logical negation: $\neg$	logical conjunction: $\wedge$	logical disjunction: $\vee$

the “truth” connective: $\top$	the “falsity” connective: $\perp$

the material conditional (implication): $p\to q$

the biconditional connective: $p \leftrightarrow q$

An implication with a false antecedent is called vacuously true.
An implication with a true consequent is called trivially true.

Operator precedence for propositional logic: $\neg\quad\wedge\quad\vee\quad\to\quad\leftrightarrow$

All operators are right-associative.
$$
\begin{array}{c|c}
\text{connective}&\text{negation}\\\hline
p\wedge q&\neg p \vee \neg q,p\to\neg q\\\hline
p\vee q&\neg p \and \neg q\\\hline
p\to q&p\wedge \neg q\\\hline
p\leftrightarrow q&p\leftrightarrow\neg q,\neg p\leftrightarrow q
\end{array}
$$

##### Lectures 4, 5: first-order logic

Unlike propositional variables, constant symbols refer to objects, not propositions.

Predicates take objects as arguments and evaluate to true or false.

the predicate =: Equality can only be applied to objects.

Functions take objects as input and produce objects as output.
	Functions evaluate to objects, not propositions.

the existential quantifier: $\exists$

the universal quantifier: $\forall$

Universally-quantified statements are said to be vacuously true in empty worlds.

“Some $P$ is a $Q$” translates as $\exists x. (P(x) \wedge Q(x))$

“All $P$s are $Q$s” translates as $\forall x. (P(x) \to Q(x))$

“All $A$s are $B$s”	$\forall x. (A(x) \to B(x))$

“Some $A$s are $B$s”	$\exists x. (A(x) \wedge B(x))$

“No $A$s are $B$s”	$\forall x. (A(x) \to \neg B(x))$

“Some $A$s aren’t $B$s”	$∃x. (A(x) \wedge \neg B(x))$

Negations:

$\forall x. P(x)$	vs	$\exists x. \neg P(x)$

$\exists x. P(x)$	vs	$\forall x. \neg P(x)$
